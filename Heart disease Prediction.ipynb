```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
df = pd.read_csv('heart.csv')

# Check and clean column names
df.columns = df.columns.str.strip().str.lower()  # Clean column names (remove spaces and convert to lowercase)
print("Columns in the dataset:", df.columns)  # Verify column names

# Define categorical columns
categorical_columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

# Verify if categorical columns exist in the dataset
missing_columns = [col for col in categorical_columns if col not in df.columns]
if missing_columns:
    print("Error: The following columns are missing in the dataset:", missing_columns)
    print("Available columns in the dataset:", df.columns)
    raise ValueError("Please check the column names in your dataset.")
else:
    print("All categorical columns are present in the dataset.")

#  Perform one-hot encoding
df = pd.get_dummies(df, columns=categorical_columns)

standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

missing_columns = [col for col in columns_to_scale if col not in df.columns]
if missing_columns:
    print("Error: The following columns are missing in the dataset:", missing_columns)
    print("Available columns in the dataset:", df.columns)
    raise ValueError("Please check the column names in your dataset.")
else:
    print("All numerical columns are present in the dataset.")

df[columns_to_scale] = standardScaler.fit_transform(df[columns_to_scale])

if 'target' not in df.columns:
    print("Error: The 'target' column is missing in the dataset.")
    print("Available columns in the dataset:", df.columns)
    raise ValueError("Please ensure your dataset has a 'target' column.")

y = df['target']
X = df.drop(['target'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

print("Data preprocessing completed successfully!")
```

```python
y = df['target']
X = df.drop(['target'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 0)
```

```python
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    knn_classifier.fit(X_train, y_train)
    knn_scores.append(knn_classifier.score(X_test, y_test))
```

```python
plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
plt.show()
```

```python
print("The score for K Neighbors Classifier is {}% with {} nieghbors.".format(knn_scores[7]*100, 8))
```

```python
svc_scores = []
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for i in range(len(kernels)):
    svc_classifier = SVC(kernel = kernels[i])
    svc_classifier.fit(X_train, y_train)
    svc_scores.append(svc_classifier.score(X_test, y_test))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for demonstration

# Use a colormap from matplotlib
colors = plt.cm.viridis(np.linspace(0, 1, len(kernels)))

# Create the bar plot
plt.bar(kernels, svc_scores, color=colors)

# Add text annotations for each bar
for i in range(len(kernels)):
    plt.text(i, svc_scores[i], f"{svc_scores[i]:.2f}", ha='center', va='bottom')

# Set labels and title
plt.xlabel('Kernels')
plt.ylabel('Scores')
plt.title('Support Vector Classifier scores for different kernels')

# Show the plot
plt.show()
```

```python
print("The score for Support Vector Classifier is {}% with {} kernel.".format(svc_scores[0]*100, 'linear'))
```

```python
dt_scores = []
for i in range(1, len(X.columns) + 1):
    dt_classifier = DecisionTreeClassifier(max_features = i, random_state = 0)
    dt_classifier.fit(X_train, y_train)
    dt_scores.append(dt_classifier.score(X_test, y_test))
```

```python
plt.plot([i for i in range(1, len(X.columns) + 1)], dt_scores, color = 'green')
for i in range(1, len(X.columns) + 1):
    plt.text(i, dt_scores[i-1], (i, dt_scores[i-1]))
plt.xticks([i for i in range(1, len(X.columns) + 1)])
plt.xlabel('Max features')
plt.ylabel('Scores')
plt.title('Decision Tree Classifier scores for different number of maximum features')
plt.show()
```

```python
print("The score for Decision Tree Classifier is {}% with {} maximum features.".format(dt_scores[17]*100, [2,4,18]))
```

```python
rf_scores = []
estimators = [10, 100, 200, 500, 1000]
for i in estimators:
    rf_classifier = RandomForestClassifier(n_estimators = i, random_state = 0)
    rf_classifier.fit(X_train, y_train)
    rf_scores.append(rf_classifier.score(X_test, y_test))
```

```python
colors = plt.cm.viridis(np.linspace(0, 1, len(estimators)))
plt.bar([i for i in range(len(estimators))], rf_scores, color = colors, width = 0.8)
for i in range(len(estimators)):
    plt.text(i, rf_scores[i], rf_scores[i])
plt.xticks(ticks = [i for i in range(len(estimators))], labels = [str(estimator) for estimator in estimators])
plt.xlabel('Number of estimators')
plt.ylabel('Scores')
plt.title('Random Forest Classifier scores for different number of estimators')
plt.show()
```

```python
print("The score for Random Forest Classifier is {}% with {} estimators.".format(rf_scores[1]*100, [100, 500]))
```
